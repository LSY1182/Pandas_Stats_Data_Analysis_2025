# **y = b + <font color = 'red'> a</font>*x**

Varing the value of a slope (aka. a)

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values
x = np.linspace(-10, 10, 100)  # Generates 100 points between -10 and 10

# Calculate the corresponding y values
y = 4 + 2 * x

# Plot the graph
plt.plot(x, y, label='y = 4 + 2x')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph with an intercept of 4 and a slope of 2')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values
x = np.linspace(-10, 10, 100)  # Generates 100 points between -10 and 10

# Calculate the corresponding y values
y = 4 - 2 * x

# Plot the graph
plt.plot(x, y, label='y = 4 - 2x')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph with an intercept of 4 and a slope of -2')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Display the plot
plt.show()
```
